# Distribution Plots in Data Visualization

We use distribution plots to understand how data is spread or distributed.  
Distribution plots help us see the frequency, pattern, shape, and spread of data values.

Note: Distribution plots are mainly used for **quantitative (numerical) data**, not qualitative (categorical) data.

Let us discuss different types of distribution visualization:

---

## 1. Histogram (Dist Plot / Hist Plot)

A histogram shows how numerical data is distributed by dividing data into bins (ranges).

- It shows frequency of data in each range.
- Helps identify skewness (left/right), normal distribution, or outliers.
- Very useful for understanding overall shape of data.

Example: Distribution of marks, age, salary, etc.

---

## 2. Rug Plot

A rug plot displays small vertical lines on the axis representing individual data points.

- Each line represents one observation.
- It is usually added below a histogram or KDE plot.
- Helps in seeing density and exact data positions.

Example: Checking how values are concentrated in a dataset.

---

## 3. Joint Plot

A joint plot shows the relationship between two numerical variables.

- It combines:
  - Scatter plot (relationship between two variables)
  - Histogram (distribution of each variable)
- Helps understand correlation and distribution together.

Example: Relationship between height and weight.

---

## 4. Pair Plot

A pair plot shows pairwise relationships between multiple numerical variables in a dataset.

- It creates scatter plots for every combination of variables.
- Diagonal shows distribution (histogram or KDE).
- Useful for exploring dataset before building machine learning models.

Example: Exploring relationships in Iris dataset.

---

## Summary

- Histogram → Distribution of single numerical variable.
- Rug Plot → Shows individual data points.
- Joint Plot → Relationship between two variables with distributions.
- Pair Plot → Relationships between multiple variables.

These plots are very useful in Exploratory Data Analysis (EDA).

```python
# sns.load_dataset('tips')

# sns = seaborn library
# load_dataset() = built-in function to load sample datasets
# 'tips' = name of dataset available inside seaborn

# This command loads the "tips" dataset.
# The tips dataset contains information about:
# total bill, tip amount, gender, day, time, size, etc.

# It returns a pandas DataFrame.
# We usually store it in a variable like this:

import seaborn as sns

df=sns.load_dataset('tips')  # Load sample tips dataset
df['size'].unique()
```

```python
# Create distribution plot (histogram)
sns.displot(df['total_bill'], bins=30, kde=True)

# df['total_bill'] → Selects total_bill column
# bins=30 → Divides data into 30 groups (bars)
# kde=True → Adds smooth curve to show data distribution pattern
```

**Joint Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a joint plot to visualize the relationship between
# total bill amount and tip amount.

sns.jointplot(
    x='total_bill',
    y='tip',
    data=df,
    kind='scatter'
)

plt.show()
```

**Pair Plot**

```python
# Pair Plot:
# Displays pairwise relationships between all numerical variables.
# Diagonal plots show the distribution of each variable.
# Off-diagonal plots show scatter plots between variable pairs.

sns.pairplot(df, hue="size", palette="rainbow")

plt.show()
```

**Rub Plot**

```python
sns.rugplot(df)
```
