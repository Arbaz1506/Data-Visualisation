### LM Plot (`lmplot`)

An **LM Plot (Linear Model Plot)** is used to visualize the relationship between two numerical variables along with a **best-fit regression line**. It helps determine whether a positive, negative, or no correlation exists between the variables.

The regression line shows the overall trend of the data, making it easier to understand how one variable changes with another.

**Quick Revision:**
- Shows relationship between two numerical variables.
- Includes a regression (best-fit) line.
- Helps identify trends and correlations.
- Useful for predictive analysis and pattern detection.

**Memory Trick:**
LM Plot = **"Scatter Plot + Best Fit Line"**

```python
import matplotlib.pyplot as plt 
import seaborn as sns
tips=sns.load_dataset('tips')
tips
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.show()
```

```python
sns.lmplot(
    x='total_bill',
    y='tip',
    data=tips,
    hue='smoker',
    palette='rainbow',
    markers=['o', 'v'],
    height=6,
    aspect=1.5,
    scatter_kws={'s': 80, 'alpha': 0.7},
    line_kws={'linewidth': 2}
)
plt.show()
```
