```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Heatmap

A **Heatmap** is a graphical representation of data where values are represented using colors. It is commonly used to visualize correlation matrices and identify patterns, trends, and relationships between variables. Different colors indicate different value ranges, making it easy to spot strong positive, negative, or weak correlations at a glance.

**Quick Revision:**
- Uses colors to represent values.
- Commonly used for correlation analysis.
- Helps identify patterns and relationships between variables.
- Darker or stronger colors usually indicate stronger relationships.

**Memory Trick:**
Heatmap = **"Colors show the strength of values or relationships."**

```python
flights= sns.load_dataset('flights')
flights
```

```python
tips= sns.load_dataset('tips')
tips
```

```python
tipcorr = tips[['total_bill', 'tip', 'size']]
tipcorr
```

```python
tipcorr.corr()
```

```python
sns.heatmap(tipcorr.corr())
```

### Cluster Map

A **Cluster Map** is an advanced version of a Heatmap that not only displays values using colors but also groups similar rows and columns together using hierarchical clustering. This makes it easier to identify patterns, relationships, and clusters within the data.

Unlike a Heatmap, a Cluster Map automatically rearranges the data so that similar observations appear close to each other.

**Quick Revision:**
- Heatmap + Automatic Clustering.
- Groups similar rows and columns together.
- Helps discover hidden patterns in data.
- Useful for correlation and pattern analysis.

**Memory Trick:**
Clustermap = **"Heatmap that automatically groups similar data together."**

```python
sns.clustermap(tipcorr.corr())
```

### Pivot Table Heatmap

A **Pivot Table Heatmap** combines a Pivot Table with a Heatmap to visualize summarized data using colors. The pivot table organizes data into rows and columns, while the heatmap uses colors to show the magnitude of values.

It helps compare values across categories and quickly identify patterns, trends, and high or low values.

**Quick Revision:**
- Pivot Table → Summarizes data into rows and columns.
- Heatmap → Uses colors to represent values.
- Pivot Table + Heatmap → Easy visualization of summarized data.
- Useful for comparing categories and finding patterns.

**Memory Trick:**
Pivot Table Heatmap = **"Summarized data shown with colors."**

```python
pvtflight = flights.pivot_table(
    values='passengers',
    index='month',
    columns='year'
)

pvtflight
```

# Matrix Plots Summary

### Heatmap
Uses colors to represent values in a matrix. Commonly used to visualize correlations and identify patterns between variables.

### Clustermap
An advanced heatmap that automatically groups similar rows and columns using hierarchical clustering, making patterns easier to discover.

### Pivot Table Heatmap
A heatmap created from a pivot table. The pivot table summarizes data into rows and columns, while the heatmap uses colors to visualize the summarized values.

## Quick Revision

- **Heatmap** → Colors show values or correlations.
- **Clustermap** → Heatmap + Automatic grouping of similar data.
- **Pivot Table Heatmap** → Summarized data displayed using colors.

## Memory Trick

**Heatmap = Colors | Clustermap = Colors + Grouping | Pivot Heatmap = Summary + Colors**
