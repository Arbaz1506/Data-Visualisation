```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Learning Basics of Matplotlib

```python
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("1st Graph")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
```

**Subplot in Matplotlib**

```python
# Create a grid of subplots with 2 rows and 2 columns
# The last number (1) means this plot will be placed in position 1 (top-left)
plt.subplot(2, 2, 1)

# Plot y values against x values in the first subplot
plt.plot(x, y)


# Create the second subplot at position 2 (top-right)
plt.subplot(2, 2, 2)

# Plot x values against y values (axes are swapped)
plt.plot(y, x)


# Create the third subplot at position 3 (bottom-left)
plt.subplot(2, 2, 3)

# Plot squared x values against squared y values
# This changes the scale and shape of the curve
plt.plot(x**2, y**2)


# Create the fourth subplot at position 4 (bottom-right)
plt.subplot(2, 2, 4)

# Plot squared x values against squared x values
# This results in a straight line since both axes use the same values
plt.plot(x**2, x**2)


# Adjust spacing between subplots to prevent overlapping
plt.tight_layout()
```

**Object Oeriented Way**

```python
# Create a new figure object (this is the main canvas)
fig = plt.figure()

# Add the first axes to the figure
# add_axes([left, bottom, width, height])
# All values are in relative figure coordinates (0 to 1)
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Plot x vs y on the first axes
axis1.plot(x, y)

# Add a second axes on top of the first one
# This creates an overlapping plot
axis2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

# Plot y vs x on the second axes
axis2.plot(y, x)
```
