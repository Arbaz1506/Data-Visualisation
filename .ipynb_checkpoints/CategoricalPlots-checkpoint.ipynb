# Categorical Data Plots

Categorical plots are used to analyze data that belongs to different categories such as Gender, Day, Department, City, etc.

### 1. Box Plot (`boxplot`)
Shows the distribution of data using quartiles and helps identify outliers.

**Use:** Compare data spread across categories.

### 2. Violin Plot (`violinplot`)
Similar to a box plot but also shows the data distribution shape.

**Use:** Understand how values are distributed.

### 3. Strip Plot (`stripplot`)
Shows each individual data point.

**Use:** Visualize all observations in a category.

### 4. Swarm Plot (`swarmplot`)
Similar to strip plot, but points do not overlap.

**Use:** Better visualization of individual observations.

### 5. Bar Plot (`barplot`)
Shows the average value of a numerical variable for each category.

**Use:** Compare category-wise averages.

### 6. Count Plot (`countplot`)
Shows the number of observations in each category.

**Use:** Count frequency of categories.

## Summary

| Plot | Purpose |
|--------|---------|
| Box Plot | Distribution & Outliers |
| Violin Plot | Distribution Shape |
| Strip Plot | Individual Data Points |
| Swarm Plot | Individual Points without Overlap |
| Bar Plot | Category-wise Average |
| Count Plot | Category-wise Count |

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
sns.countplot(x='sex', data=df, hue='smoker')
plt.show()
```

**Bar plot**

```python
# Bar Plot
# Shows the average total_bill for each gender.
# Height of each bar represents the mean value.
# Useful for comparing averages across categories.

sns.barplot(x='sex', y='total_bill', data=df)
plt.show()
```

**Box Plot**

```python
# Box Plot
# Shows the distribution of total_bill for each gender.
# The line inside the box represents the median.
# The box represents the middle 50% of the data.
# Points outside the whiskers are potential outliers.

sns.boxplot(x='tip', y='day', data=df)
plt.show()
```

### Box Plot and Violin Plot

Both **Box Plot** and **Violin Plot** are used to understand the distribution of numerical data across different categories.

A **Box Plot** shows the median, quartiles, spread of the data, and possible outliers. It provides a quick summary of the data and is useful for comparing distributions and detecting unusual values.

A **Violin Plot** contains all the information of a box plot but also displays the density of the data. Wider sections indicate a higher concentration of observations, while narrower sections indicate fewer observations. This helps in understanding the shape of the distribution.

**Quick Revision:**
- **Box Plot** → Shows spread, median, quartiles, and outliers.
- **Violin Plot** → Shows spread, median, quartiles, outliers, and distribution shape.
- Use **Box Plot** for a simple summary.
- Use **Violin Plot** when you want to see how the data is distributed.

**Violin plot**

```python
sns.violinplot(x="day",y="total_bill", data=df)
```

### Strip Plot and Swarm Plot

Both **Strip Plot** and **Swarm Plot** are used to display individual data points for different categories. They help visualize the distribution of data and identify clusters or outliers.

A **Strip Plot** displays all data points, but points may overlap when many observations have similar values. It is simple and useful for small datasets.

A **Swarm Plot** is an improved version of the strip plot. It adjusts the position of points so that they do not overlap, making the distribution easier to understand and individual observations more visible.

**Quick Revision:**
- **Strip Plot** → Shows all data points, but points may overlap.
- **Swarm Plot** → Shows all data points without overlapping.
- Use **Strip Plot** for a quick view of observations.
- Use **Swarm Plot** when you want a clearer visualization of individual data points.

```python
sns.stripplot(x="day",y="total_bill", data=df)
```

```python
sns.swarmplot(x="day",y="total_bill", data=df)
```

# Categorical Plots Summary

### Box Plot
Shows the distribution of data using median, quartiles, spread, and outliers.

### Violin Plot
Shows the distribution and density (shape) of the data along with box plot information.

### Strip Plot
Displays all individual data points, but points may overlap.

### Swarm Plot
Displays all individual data points without overlapping, making the distribution clearer.

### Bar Plot
Shows the average (mean) value of a numerical variable for each category.

### Count Plot
Shows the number of observations present in each category.

## Quick Revision

- **Box Plot** → Distribution + Outliers
- **Violin Plot** → Distribution + Density Shape
- **Strip Plot** → Individual Points (may overlap)
- **Swarm Plot** → Individual Points (no overlap)
- **Bar Plot** → Category-wise Average
- **Count Plot** → Category-wise Count
